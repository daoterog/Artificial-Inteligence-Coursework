```python
from scipy.stats import binom
import numpy as np
import matplotlib.pyplot as plt
```

```python
def binomial_shannon_entropy(n_size: int, prob: float) -> tuple[list, list, list]:

    shannon_entropy = 0
    probabilities = []
    cumulative_probabilities = [0]
    ks = []

    for k in range(0, n_size):
        p_k = binom.pmf(k, n_size, prob)
        shannon_entropy += -p_k*np.log(p_k)

        ks.append(k)
        probabilities.append(shannon_entropy)
        cumulative_probabilities.append(cumulative_probabilities[-1] + shannon_entropy)

    return ks, probabilities, cumulative_probabilities

def plot_shannon_entropy(ks: list, probabilities: list, cumulative_probabilities: list) -> None:

    _, ax = plt.subplots(1, 2, figsize=(15, 5))
    ax[0].plot(ks, probabilities)
    ax[0].set_title("Probability Mass Functions")
    ax[0].set_xlabel('k')
    ax[0].set_ylabel('probabilities')

    ax[1].plot(ks, cumulative_probabilities[1:])
    ax[1].set_title("Cumulative Probability Distribution")
    ax[1].set_xlabel('k')
    ax[1].set_ylabel('probabilities')

entropies = []
for i in range(1, 100):
    prob = i/100
    ks, probabilities, cumulative_probabilities = binomial_shannon_entropy(1000, prob)

    entropies.append(probabilities[-1])

max(entropies)
```

```python
prob = 0.5
N = 1000
ks, probabilities, cumulative_probabilities = binomial_shannon_entropy(N, prob)

plot_shannon_entropy(ks, probabilities, cumulative_probabilities)
```
